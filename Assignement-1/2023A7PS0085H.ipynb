```python
#------Data Cleaning & Descriptive Stats (with BEFORE & AFTER plots)-----------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ---------- Load ----------
df = pd.read_csv("/content/housing.csv")
print("Shape:", df.shape)
display(df.head())

# ---------- Columns ----------
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# ---------- Descriptive stats (mean, median, std) ----------
stats = pd.DataFrame({
    "mean": df[num_cols].mean(),
    "median": df[num_cols].median(),
    "std": df[num_cols].std()
})
print("\nNumeric summary (BEFORE imputation):")
display(stats)

# ---------- Mode (first mode for each column) ----------
modes = df.mode(dropna=True)
mode_row = modes.iloc[0] if not modes.empty else pd.Series({c: np.nan for c in df.columns})
print("\nMode (first mode per column):")
display(mode_row)

# ---------- Visualize descriptive stats (bar chart) ----------
stats.plot(kind="bar", figsize=(10,5))
plt.title("Descriptive statistics (mean, median, std) — BEFORE imputation")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------- Duplicates ----------
print("\nDuplicate rows (before):", df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print("Duplicate rows (after):", df.duplicated().sum())

# ---------- Missing values BEFORE ----------------------
print("\nMissing values BEFORE imputation:")
display(df.isna().sum())

# keep a copy for BEFORE plots
df_before = df.copy()

# ---------- Imputation (numeric -> mean/median, categorical -> mode) ----------
IMPUTE_NUMERIC_WITH = "median"   # set to "mean" to use mean instead

df_clean = df.copy()

# numeric
for c in num_cols:
    if df_clean[c].isna().any():
        if IMPUTE_NUMERIC_WITH == "mean":
            df_clean[c].fillna(df_clean[c].mean(), inplace=True)
        else:
            df_clean[c].fillna(df_clean[c].median(), inplace=True)

# categorical
for c in cat_cols:
    if df_clean[c].isna().any():
        m = df_clean[c].mode(dropna=True)
        fill = m.iloc[0] if not m.empty else "Unknown"
        df_clean[c].fillna(fill, inplace=True)

# ---------- Missing values AFTER ----------
print("\nMissing values AFTER imputation:")
display(df_clean.isna().sum())

# ---------- Distributions BEFORE vs AFTER (numeric) ----------
for c in num_cols:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.histplot(df_before[c].dropna(), kde=True, bins=30)
    plt.title(f"{c} — BEFORE (non-missing)")
    plt.xlabel(c)

    plt.subplot(1,2,2)
    sns.histplot(df_clean[c], kde=True, bins=30)
    plt.title(f"{c} — AFTER (imputed)")
    plt.xlabel(c)

    plt.tight_layout()
    plt.show()

# ---------- Save cleaned dataset ----------
df_clean.to_csv("housing_cleaned.csv", index=False)
print("\n Cleaned dataset saved as 'housing_cleaned.csv'.")


# ----- Descriptive Stats & Basic Visualizations (using df_clean from Part 1)-----

# Histograms
for c in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df_clean[c], bins=30, kde=True)
    plt.title(f"Distribution of {c}")
    plt.xlabel(c)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Scatter plot: longitude vs latitude (color = median_house_value)
if all(col in df_clean.columns for col in ["longitude", "latitude", "median_house_value"]):
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(
        df_clean["longitude"], df_clean["latitude"],
        c=df_clean["median_house_value"], cmap="viridis", alpha=0.5
    )
    plt.colorbar(scatter, label="Median House Value")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.title("Geographical Distribution of Houses by Value")
    plt.show()

#  Boxplots overview
plt.figure(figsize=(12,6))
sns.boxplot(data=df_clean[num_cols], orient="h")
plt.title("Boxplots of Numerical Features")
plt.tight_layout()
plt.show()


# -------------Exploratory Visualizations ----------------------------

#  Grouping by ocean_proximity and summary statistics
grouped = df_clean.groupby('ocean_proximity')
print("Summary statistics by ocean_proximity:")
display(grouped[['median_house_value', 'median_income', 'total_rooms']].mean().round(2))

#  Visualizing target variable distribution
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.boxplot(x='ocean_proximity', y='median_house_value', data=df_clean)
plt.title('House Value by Ocean Proximity')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
avg_values = df_clean.groupby('ocean_proximity')['median_house_value'].mean()
avg_values.plot(kind='bar', color='skyblue')
plt.title('Average House Value by Ocean Proximity')
plt.xticks(rotation=45)
plt.ylabel('Average House Value')

plt.tight_layout()
plt.show()


#  other observations
large_rooms = df_clean[df_clean['total_rooms'] >= 1500]
print(f"Houses with total_rooms >= 1500: {large_rooms.shape}")

small_bedrooms = df_clean[df_clean['total_bedrooms'] < 1500]
high_income = df_clean[df_clean['median_income'] > 10]
print(f"Houses with <1500 bedrooms: {small_bedrooms.shape}")
print(f"Houses with income >10: {high_income.shape}")

# Premium houses analysis
premium_houses = df_clean[df_clean['median_house_value'] > df_clean['median_house_value'].quantile(0.8)]
print(f"• Premium houses (top 20%): {premium_houses.shape[0]:,}")
print("  Premium houses by proximity:")
premium_by_proximity = premium_houses['ocean_proximity'].value_counts()
for category, count in premium_by_proximity.items():
    pct = count / len(premium_houses) * 100
    print(f"    - {category}: {count} ({pct:.1f}%)")

# Label encoding
from sklearn.preprocessing import LabelEncoder

ocean_proximity_le = LabelEncoder()
df_clean['ocean_proximity'] = ocean_proximity_le.fit_transform(df_clean['ocean_proximity'])

print("\nEncoding mapping:")
for i, category in enumerate(ocean_proximity_le.classes_):
    print(f"{category} -> {i}")

print("\nSample encoded data:")
display(df_clean[['ocean_proximity']].head())

# insights
highest_value_category = grouped['median_house_value'].mean().idxmax()
lowest_value_category = grouped['median_house_value'].mean().idxmin()
price_diff = grouped['median_house_value'].mean().max() - grouped['median_house_value'].mean().min()

print(f"\n Key Insight:")
print(f"• Highest value area: {highest_value_category}")
print(f"• Lowest value area: {lowest_value_category}")
print(f"• Price difference: ${price_diff:,.0f} ({price_diff/grouped['median_house_value'].mean().min()*100:.0f}% premium)")



#-------Correlation and Premium Analysis--------------------------------------

sns.set(style="whitegrid")

target = "median_house_value"
num_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()

# 11(a). Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df_clean[num_cols].corr(numeric_only=True), annot=True, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()

# 11(b) Boxplots
sns.boxplot(data=df_clean, x="total_rooms", palette="viridis")
plt.title("Boxplot of total_rooms")
plt.show()

# 11(c) Insights from correlation and boxplots
rows = []
N = len(df_clean)
for c in [f for f in num_cols if f != target]:
    s = df_clean[c].dropna()
    corr_val = s.corr(df_clean[target])
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    low, high = q1 - 1.5*iqr, q3 + 1.5*iqr
    out_count = ((s < low) | (s > high)).sum()
    rows.append((c, corr_val, out_count))

metrics = pd.DataFrame(rows, columns=["feature", "corr_with_target", "outlier_count"]).set_index("feature")

print("\nTop positive correlations with target:")
display(metrics.sort_values("corr_with_target", ascending=False).head(3))

print("\nTop negative correlations with target:")
display(metrics.sort_values("corr_with_target").head(3))

print("\nFeatures with most outliers:")
display(metrics.sort_values("outlier_count", ascending=False).head(5))

# 11(d) + (e) Premium Houses Analysis
threshold = df_clean[target].quantile(0.90)
df_clean["is_premium"] = df_clean[target] >= threshold
print(f"Premium threshold = ${threshold:,.0f}; Premium houses = {df_clean['is_premium'].sum()}")

df_clean["value_per_room"] = df_clean[target] / df_clean["total_rooms"].clip(lower=1)
df_clean["price_to_income"] = df_clean[target] / df_clean["median_income"].clip(lower=0.1)

plt.figure(figsize=(8,5))
sns.boxplot(data=df_clean, x="ocean_proximity", y=target, palette="viridis")
plt.title("House Value by Ocean Proximity")
plt.xticks(rotation=45)
plt.show()

premium_df = df_clean[df_clean["is_premium"]]

for col in ["value_per_room", "price_to_income", "median_income"]:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=premium_df, x="ocean_proximity", y=col, palette="viridis")
    plt.title(f"Premium Houses: {col} by Ocean Proximity")
    plt.xticks(rotation=45)
    plt.show()

summary = premium_df.groupby("ocean_proximity")[
    [target, "value_per_room", "median_income"]
].median().round(2)

print("\nMedian values for premium houses (by proximity):")
display(summary)


# ------Advanced Analysis & Feature Engineering ------------------------

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Ensure df_clean exists from previous steps
print("Dataset shape for Member 5:", df_clean.shape)

#12(a) KDE + Boxplot + Scatterplot Function
def plot_feature_distribution(df, feature, target="median_house_value"):
    """Plots KDE, boxplot, and scatterplot for a given numeric feature."""
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # KDE
    df[feature].plot(kind="kde", ax=axes[0])
    axes[0].set_title(f"KDE of {feature}")

    # Boxplot
    axes[1].boxplot(df[feature].dropna())
    axes[1].set_title(f"Boxplot of {feature}")
    axes[1].set_xticks([])

    # Scatter with target
    axes[2].scatter(df[feature], df[target], alpha=0.3)
    axes[2].set_xlabel(feature)
    axes[2].set_ylabel(target)
    axes[2].set_title(f"{feature} vs {target}")

    plt.tight_layout()
    plt.show()

# Example: plot for median_income
plot_feature_distribution(df_clean, "median_income")

# 12(b) Quartile-based KDE + Strip plots
for col in ["median_income", "total_rooms", "housing_median_age"]:
    q1, q2, q3 = df_clean[col].quantile([0.25, 0.5, 0.75])
    print(f"\nQuartiles for {col}: Q1={q1:.2f}, Q2={q2:.2f}, Q3={q3:.2f}")

    # KDE
    df_clean[col].plot(kind="kde", figsize=(6,4), title=f"KDE of {col}")
    plt.show()

    # Strip plot by quartile bins
    bins = [df_clean[col].min(), q1, q2, q3, df_clean[col].max()]
    labels = ["Q1", "Q2", "Q3", "Q4"]
    df_clean[f"{col}_quartile"] = pd.cut(df_clean[col], bins=bins, labels=labels, include_lowest=True)

    plt.figure(figsize=(7,4))
    plt.scatter(df_clean[f"{col}_quartile"], df_clean["median_house_value"], alpha=0.3)
    plt.title(f"Median House Value by {col} Quartiles")
    plt.xlabel(f"{col} Quartile")
    plt.ylabel("Median House Value")
    plt.show()

# 12(c) Outlier Detection (±2 std deviation)
outlier_counts = {}
for col in df_clean.select_dtypes(include=[np.number]).columns:
    mean, std = df_clean[col].mean(), df_clean[col].std()
    lower, upper = mean - 2*std, mean + 2*std
    outliers = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()
    outlier_counts[col] = outliers

outlier_df = pd.DataFrame.from_dict(outlier_counts, orient="index", columns=["outlier_count"])
print("\nOutlier counts (±2 std):")
display(outlier_df.sort_values("outlier_count", ascending=False).head(10))

#12(d) Feature Scaling
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scalers = {
    "StandardScaler": StandardScaler(),
    "MinMaxScaler": MinMaxScaler()
}

scaled_features = {}
for name, scaler in scalers.items():
    scaled = scaler.fit_transform(df_clean[num_cols])
    scaled_features[name] = pd.DataFrame(scaled, columns=num_cols)
    print(f"\n{name} applied. Example:")
    display(scaled_features[name].head())

# 12(e) New Feature Creation
df_clean["rooms_per_household"] = df_clean["total_rooms"] / df_clean["households"].replace(0, np.nan)
df_clean["bedrooms_ratio"] = df_clean["total_bedrooms"] / df_clean["total_rooms"].replace(0, np.nan)

# Visualize new features
for col in ["rooms_per_household", "bedrooms_ratio"]:
    plt.figure(figsize=(6,4))
    plt.hist(df_clean[col].dropna(), bins=30, alpha=0.7)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

print("\nNew features added: 'rooms_per_household', 'bedrooms_ratio'")

```
